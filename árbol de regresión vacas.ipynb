# Imports

```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
```

# Leer el dataset

```python
data = pd.read_csv("Downloads/agroindustria_bd15f73c-fe07-41d9-9dd9-58e3244cad59.csv")
data
```

# Ajustar columnas dataset para poder hacer la predicción

```python
colnames = data.columns.values.tolist()
predictors = colnames[31:32]
target = colnames[22:23]
X = data[target]
Y = data[predictors]
```

# Condiciones árbol y ajuste a X,Y

```python
regtree = DecisionTreeRegressor(min_samples_split=2, min_samples_leaf=2, max_depth=15, random_state=0)
```

```python
regtree.fit(X,Y)
```

# Predicción

```python
preds = regtree.predict(data[predictors])
```

```python
data["preds"] = preds
```

```python
data[["preds", "Precio - MEJ"]] # aquí se compara la prediccionde datos con los datos originales
```
